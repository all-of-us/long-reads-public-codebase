```python
original_filename = "/home/jupyter/HG002_sv_cleaned.vcf.gz"
phased_file = "/home/jupyter/HG002_phased_sv.sorted.vcf.gz"
```

```python
with gzip.open(original_filename, 'r') as fp:
    original = fp.readlines()
```

```python
Original = {}
for line in original:
    line = line.decode()
    if line.startswith("#"):
        continue
    itemlist = line.split('\t')
    pos = itemlist[1]
    genotype = itemlist[9].split(':')[0].split("/")
    Original[pos] = genotype
```

```python
with gzip.open(phased_file, 'r') as fp:
    phased = fp.readlines()
```

```python
for line in phased:
    line = line.decode()
    if line.startswith("#"):
        continue
    itemlist = line.split('\t')
    pos = itemlist[1]
    genotype = itemlist[9].split(':')[0]
    if "/" in genotype:
        g = genotype.split('/')
    else:
        g = genotype.split('|')
    if sorted(g) != sorted(Original[pos]):
        print(pos, g, Original[pos])
```
