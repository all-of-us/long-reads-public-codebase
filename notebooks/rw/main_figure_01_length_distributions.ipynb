```python
import numpy as np
from collections import Counter
import os
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
if not os.path.exists('full_report_map.txt'):
    !gsutil -u $GOOGLE_PROJECT cat gs://fc-aou-datasets-controlled/pooled/longreads/v7_base/hifiasm/*/report_map.txt > full_report_map.txt
```

```python
hap1_n50 = !grep '^N50' full_report_map.txt | awk '{ print $3 }'
hap2_n50 = !grep '^N50' full_report_map.txt | awk '{ print $4 }'
```

```python
hap1_n50 = [int(a) for a in hap1_n50]
hap2_n50 = [int(a) for a in hap2_n50]
```

```python
plt.figure(figsize=(10, 6))
plt.hist(hap1_n50, bins=range(int(min(hap1_n50)), int(max(hap1_n50)) + 100, 100), color='blue', alpha=0.7)
plt.title('Histogram of Genome Assembly N50 Values')
plt.xlabel('N50 (bp)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)

plt.show()
```
